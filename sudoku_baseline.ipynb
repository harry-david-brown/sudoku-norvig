```python
def display_sudoku(puzzle_string):
    """Display an 81-character sudoku string as a readable 9x9 grid."""
    assert len(puzzle_string) == 81, "Puzzle must be 81 characters"
    
    for i, char in enumerate(puzzle_string):
        # Replace 0 with dot for readability
        val = '.' if char == '0' else char
        print(val, end=' ')
        
        # Vertical box separator
        if (i + 1) % 3 == 0 and (i + 1) % 9 != 0:
            print('|', end=' ')
        
        # End of row
        if (i + 1) % 9 == 0:
            print()
        
        # Horizontal box separator
        if (i + 1) % 27 == 0 and (i + 1) != 81:
            print('------+-------+------')

# Test it
puzzle = "004300209005009001070060043006002087190007400050083000600000105003508690042910300"
display_sudoku(puzzle)
```
